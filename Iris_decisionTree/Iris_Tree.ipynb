```python
import pydotplus
from sklearn.datasets import load_iris
from sklearn import tree
from IPython.display import Image

#load the iris dataset in
iris = load_iris()

#create a Decision Tree Classifier
clf = tree.DecisionTreeClassifier()

#Provide the tree with data for supervised learning
clf = clf.fit(iris.data, iris.target)

#get tree to predict.
print clf.predict([[5.1, 3.5, 1.4, 0.2]])

#Create image of tree
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names,
                                filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

Checking tree accuracy:

```python
from sklearn.datasets import load_iris
from sklearn import tree

#load the iris dataset in
iris = load_iris()

x = iris.data
y = iris.target

#Splitting the dataset in half so that one half can train the classifier and the other to test
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.5)

#create a Decision Tree Classifier
clf = tree.DecisionTreeClassifier()

#Provide the tree with data for supervised learning
clf = clf.fit(x_train, y_train)

#Getting the Classifier to predict the half of data we took away
predictions = clf.predict(x_test)

#Checking accuracy by comparing the predictions to the real data
from sklearn.metrics import accuracy_score
print accuracy_score(y_test, predictions)
```
